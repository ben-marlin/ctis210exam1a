# Basics of Coding

We'll launch into using a CAS shortly, but there are some names & conventions you'll need to make sense of it all. The first is the notion of importing libraries - this just means there are external resources that our programs will invoke. But we have to tell the program to do so! For instance, if you want to use the tool `arange()` from the `numpy` package, you need to do this:

`from numpy import arange`

You should always do this at the start of your Jupyter notebook cell, and I would advise you to include the import in *every* cell. That way they can be run in any order. However, it may be that you need to import a large number of features from the same package. In this case, use

`from numpy import *`

Lastly, you may need to import a numpy feature that duplicates a name from another package. This appears to be very common, as many example code snippets include something like this

`import numpy as np`

`import math as m`

`numpy_tau = 2 * np.pi`

`math_tau = 2 * m.pi`

This allows you to have two different versions of $\pi$ or $\sqrt{\cdot}$, or whatever. Shouldn't come up very often for us, but when you see it online, I don't want you to be startled by it!


```python
from numpy import arange

x = arange(10)

y = x[5:]

z = x[:5]

aa = x[::2]

bb = x[::-1]

print(x, '\n', y, '\n',z, '\n',aa, '\n',bb)
```

The preceding is a code cell. You should be able to figure out how to insert both code and text cells. I left the results after it, which you can delete by clicking over the rectangle & arrow. As you should guess, the play button on the cell runs it again and reproduces the results. Alternately - and I highly suggest you get used to doing this - while your cursor is in the cell, you can run it by hitting Control-Enter or Command-Enter.

We created the tags `x, y, z, aa, bb` here and assigned them to objects created from the `arange` function. (I will try to be consistent about calling these tags rather than variables, as a tag doesn't actually vary!)

You should be able to deduce what the print statement does, but the part with brackets is pretty arcane. The `\\n` produces a newline character. Note that `x` is an array that starts with 0, but ends just before reaching 10. It does, however, have 10 elements in it.

The other variables are slices of `x`. `y` has cut off the entries that preceded the 5. `z` cut off the elements starting at 5. `aa` cut out elements by skipping over 2 at a time. Finally, `bb` reversed the order of `x`, which feels like it doesn't fit the slice metaphor.

```python
from numpy import arange

x = arange(4,20)

y = x[:5]

z = y * 2

aa = z - 3

bb = aa / 2

cc = aa // 2

print(x, y, z, aa, bb, cc, x[2], z * aa, sep = "\n")
```

Note here that `numpy` arrays can have arithmetic operations done to them, and when this results in non-integers, it expresses in decimal. If you want it to drop the fractional part, us the double divide to specify integer division. Note also that when arrays of the same size are multiplied, it does this term-by-term. If the arrays are different sizes, an error occurs.

# Variables in SymPy

Next, we'll introduce something that is more properly a variable, using the CAS package `sympy`. These will allow us to do some algebra, graphing, and eventually calculus.

```python
from sympy import var, expand, factor

x = var('x')      # it's a good idea to use the same name for the tag and variable!

expand((2*x-3)*(3*x+4))
```

```python
from sympy import var, expand, factor

x = var('x')

print(expand(x*(x-1)))
```

```python
from sympy import var, expand, factor, pprint

x = var('x')

pprint(expand((2*x-1)*(2*x-1)))
```

Notice the 3 different print styles. Ideally, we'd like to see the first, but they should all be readable. If you'd like to experiment more with this, see what you can find on Google.

Note also the use of * for multiplication and ** for exponentiation. The notation for coding is not as nice as paper & pencil algebra!

Note also that if you're used to using ^ for exponents (like I am), in Python, that means something completely different!

<font color = red> **Use this to multiply $(x^2+4x+1)(x-3)$ and $(2x+7y)(x-3y)$ *here*.**

```python
from sympy import var, expand, factor, pprint
x = var('x')
expand ((x**2+4*x+1)*(x-3))
```

```python
from sympy import var, expand, factor, pprint
x = var('x')
y = var('y')
expand ((2*x+7*y)*(x-3*y))
```

```python
from sympy import var, expand, factor, pprint

x = var('x')

factor(x**3-8)
```

Now that you know how to make `SymPy` multiply polynomials, it would be handy to have it factor them as well. Notice it doesn't by default factor over irrational numbers.

<font color = red> **Use this idea to factor $x^4-16$ and $14x^2+29x-15$ *here*.**

```python
from sympy import var, expand, factor, pprint
x = var('x')
factor (x**4-16)
```

```python
from sympy import var, expand, factor, pprint
x = var('x')
factor (14*x**2+29*x-15)
```

```python
from sympy import plot

plot(x**3 - 12 * x)
```

Check that out - we just told it to graph the function and it made a guess what domain we wanted. It usually makes pretty good choices, but if we want a better view, we might play around with the `x` values we want.

```python
from sympy import plot

plot(x**3 - 12 * x, (x,-4,4))
```

That's a little better. If you want some more control over it, you can Google for "sympy plot" and find out how to adjust the range as well.

```python
from sympy import solve

plot(x**3 - 12 * x, 3*x, (x,-4,4))

solve((x**3 - 12 * x) - (3*x), x)
```

This is another cool feature - you can get it to do symbolic solutions for you, not just approximations!

Notice we had to import the `solve` feature. I produced a graph to help you see what it's doing.

Also notice that it's set up so that `solve(f(x),x)` really means "solve f(x) = 0 for x".

<font color = red> **Now use this to find where $y=4x-x^2$ intersects $y=x^3$.**

```python
from sympy import solve
plot (4*x-x**2, x**3)
solve ((4*x-x**2)-(x**3), x)
```
